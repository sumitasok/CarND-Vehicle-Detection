```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline

image = mpimg.imread('input/test1.jpg')
plt.imshow(image)
```

```python
image_cp = np.copy(image)
cv2.rectangle(image_cp, (1070, 400), (1250, 580), (0,0,255), 6)
cv2.rectangle(image_cp, (400, 400), (1275, 580), (0,255,0), 6)
plt.imshow(image_cp)
```

```python
image_cp = np.copy(image)
cv2.rectangle(image_cp, (800, 400), (900, 500), (0,0,255), 6)
cv2.rectangle(image_cp, (400, 400), (1275, 570), (0,255,0), 6)
plt.imshow(image_cp)
```
